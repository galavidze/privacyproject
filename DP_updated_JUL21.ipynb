```python
def split_shuffle(df):
    '''
    Shuffles the main dataframe and splits into 15 parts.
    '''
    df = shuffle(df)
    df.reset_index(inplace=True,drop=True)
    df1 = df[0:1222]
    df2 = df[1223:2445]
    df3 = df[2446:3668]
    df4 = df[3669:4891]
    df5 = df[4892:6114]
    df6 = df[6115:7337]
    df7 = df[7338:8560]
    df8 = df[8561:9783]
    df9 = df[9784:11006]
    df10 = df[11007:12229]
    df11 = df[12230:13452]
    df12 = df[13453:14676]
    df13 = df[14677:15899]
    df14 = df[15900:17122]
    df15 = df[17123:]
    return df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15
```

```python
def vaccplace_born(df):
    ''' 
    Returns vaccination place and born DP and raw datasets
    '''
    e = 1/2
    lis_city = []
    lis_city_p = []
    lis_city_dp = []
    total = 0
    df_raw_born_vaccplace = pd.DataFrame(columns = born_unique)
    df_dp_born_vaccplace = pd.DataFrame(columns = born_unique)
    for born in born_unique:
        lis_city_p = []
        lis_city = []
        lis_city_dp = []
        inc = 63
        for vaccplace in unique:
            inc -= 1
            vaccplace_len = len(df.loc[(df['vaccinationPlace'] == vaccplace) & (df['Born'] == born)])
            total += vaccplace_len
            lis_city.append(vaccplace_len)
            dp_vaccplace_len = vaccplace_len + np.random.laplace(0,1/e)
            if(dp_vaccplace_len < 0 ): dp_vaccplace_len = 0
            lis_city_dp.append(dp_vaccplace_len)
            if(inc == 0):
                lis_city_p = [round(i) for i in lis_city]
                df_raw_born_vaccplace.loc[len(df_raw_born_vaccplace)] = lis_city_p
                lis_city_p = [round(i) for i in lis_city_dp]
                df_dp_born_vaccplace.loc[len(df_dp_born_vaccplace)] = lis_city_p
                lis_city_dp = []
        total = 0
    return df_raw_born_vaccplace, df_dp_born_vaccplace
```

```python
def born_vacctype(df):
    '''
    Returns born and vaccination type DP and raw datasets.
    '''
    born_unique = u_born
    vacc_types = ['Born','Sinopharm','Pfizer','SINOVAC','AstraZeneca']
    lis_city = []
    lis_city_lap = []
    for born in born_unique:
    #Raw
        sino = int(len(df.loc[(df['Born'] == born) & (df['vaccineName'] == 'Sinopharm')]))
        pfi = int(len(df.loc[(df['Born'] == born) & (df['vaccineName'] == 'Pfizer')]))
        svac = int(len(df.loc[(df['Born'] == born) & (df['vaccineName'] == 'SINOVAC')]))
        az = int(len(df.loc[(df['Born'] == born) & (df['vaccineName'] == 'AstraZeneca')]))
        lis_city.append([born,sino,pfi,svac,az])

        #DP
        e = 0.5
    
        sino_dp = int(round(sino + np.random.laplace(0,1/e)))
        pfi_dp = int(round(pfi + np.random.laplace(0,1/e)))
        svac_dp = int(round(svac + np.random.laplace(0,1/e)))
        az_dp = int(round(az + np.random.laplace(0,1/e)))

        if(sino_dp < 0): sino_dp = 0
        if(pfi_dp < 0): pfi_dp = 0
        if(svac_dp < 0): svac_dp = 0
        if(az_dp < 0): az_dp = 0

        lis_city_lap.append([born,sino_dp,pfi_dp,svac_dp,az_dp])


    df_raw_vacctype = pd.DataFrame(columns=vacc_types)
    df_raw_vacctype = df_raw_vacctype.append(pd.DataFrame(lis_city, columns = df_raw_vacctype.columns))
    df_DP_vacctype = pd.DataFrame(columns=vacc_types)
    df_DP_vacctype = df_DP_vacctype.append(pd.DataFrame(lis_city_lap,columns = df_DP_vacctype.columns))

    return df_raw_vacctype, df_DP_vacctype
```

```python
#Preparing the final datasets
born_cols = ['Born','Sinopharm','Pfizer','SINOVAC','AstraZeneca']
df_raw_born_vacctype = pd.DataFrame(columns=born_cols)
df_dp_born_vacctype = pd.DataFrame(columns=born_cols)
df_raw_born_vacctype['Born'] = u_born
df_dp_born_vacctype['Born'] = u_born

df_raw_born_vaccplace = pd.DataFrame(columns=u_born)
df_dp_born_vaccplace = pd.DataFrame(columns=u_born)


df_raw_born_vaccplace = df_raw_born_vaccplace.fillna(0)
df_dp_born_vaccplace = df_dp_born_vaccplace.fillna(0)

df_dp_born_vacctype = df_dp_born_vacctype.fillna(0)
df_raw_born_vacctype = df_raw_born_vacctype.fillna(0)
```

```python
#run 15 times (because lotto lasted 15 weeks) 
#P.S. takes about 3-4 minutes to run on my mac.
for i in range(15):
    df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15 = split_shuffle(df) #split into 15 datasets
    df_list = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15] 
    #configuring the combined datasets
    for dataframe in df_list:
        raw_born_vaccplace, dp_born_vaccplace = vaccplace_born(dataframe) 
        raw_born_vacctype, dp_born_vacctype = born_vacctype(dataframe)

        df_dp_born_vaccplace = df_dp_born_vaccplace.add(dp_born_vaccplace,fill_value = 0)
        df_dp_born_vacctype = df_dp_born_vacctype.add(dp_born_vacctype,fill_value=0)

        df_raw_born_vaccplace = df_raw_born_vaccplace.add(raw_born_vaccplace,fill_value = 0)
        df_raw_born_vacctype = df_raw_born_vacctype.add(raw_born_vacctype,fill_value = 0)

```
